```python
import matplotlib.pyplot as plt
```

Subplot은 하나의 결과 사진에 여러개의 그래프를 넣는거다. subplot(행개수, 열개수, 이번 그래프의 인덱스)로 액자를 짜고 원하는 액자 부위를 선택하면,
그 액자 자리에서 plt.plot()으로 평범한 그래프 그리기를 하면된다.
아래에서 subplot(1,2,3)은 알다시피, 전체 나눠놓은액자 개수가 1x2 (가로x세로) 인데, 3이라는 인덱스는 있을 수 없어 에러를 유발한다.

```python
plt.subplot(1,2,1)
plt.plot([1,2,3,4],[3,4,5,6], 'ro') # r: 붉은색 o: 점
plt.subplot(1,2,2) # 이런 식으로 현재 편집할 액자를 바꾸는게 이상하지만, matplotlib의 특이한 문법으로 보인다.
plt.plot([1,2,3,4],[3,4,5,6], 'b-') # b: 파란색 -: 선
plt.axis([1,7,1,6]) # axis 함수로 축의 범위를 지정
# plt.subplot(1,2,3)
# plt.plot([1,2,3,4],[3,4,5,6], 'g-')
```

사실, matplotlib는 numpy의 array를 일반적으로 사용한다. 그래서, numpy array를 다루는 경험도 조금 있으면 좋다.

그리고 아래 그래프를 보면 알 수 있듯이, 하나의 그래프, 하나의 축 안에 여러개의 대응관계를 표현할 수도 있다.

```python
import numpy as np

t = np.arange(0., 5., 0.2)  # 200ms 간격으로 균일한 샘플된 시간

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
